```python
cust = pd.read_csv('ejr787work_cust.csv', header=0)
item = pd.read_csv('ejr787work_item.csv', header=0)
mail = pd.read_csv('ejr787work_mail.csv', header=0)
```

```python
# 2: Merge buyer_status onto the other two dataframes and then create new dataframes which only have ACTIVE customers.  
item_merge = pd.merge(item,cust[['acctno','buyer_status']],on='acctno', how='left')
mail_merge = pd.merge(mail,cust[['acctno','buyer_status']],on='acctno', how='left')

item_active = item_merge.loc[item_merge['buyer_status'] == "ACTIVE"]
mail_active = mail_merge.loc[mail_merge['buyer_status'] == "ACTIVE"]
cust_active = cust.loc[cust['buyer_status'] == "ACTIVE"]
```

```python
conn = sqlite3.connect('xyz.db')
c = conn.cursor()
```

```python
# Write all three dataframes to xyz.db and check the database to make sure it uploaded properly.  
cust_active.to_sql("customer", conn)
mail_active.to_sql("mail", conn)
item_active.to_sql("item", conn)

pd.read_sql_query("SELECT * FROM item LIMIT 5;", conn)
```

```python
cust_mod['visa'] = np.where((np.logical_or(cust_mod['visa_reg']=='Y',cust_mod['visa_prem']=='Y')), 'Y', 'N')
cust_mod['amex'] = np.where((np.logical_or(cust_mod['amex_reg']=='Y',cust_mod['amex_prem']=='Y')), 'Y', 'N')
cust_mod['disc'] = np.where((np.logical_or(cust_mod['disc_reg']=='Y',cust_mod['disc_prem']=='Y')), 'Y', 'N')
cust_mod['mast'] = np.where((np.logical_or(cust_mod['mc_reg']=='Y',cust_mod['mc_prem']=='Y')), 'Y', 'N')
```
